```
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
```

```
dir = '/content/drive/My Drive/Dataset_Object'
```

```
categories = ['Ball', 'Bat']
data = []
for category in categories:
  path = os.path.join(dir,category)
  label = categories.index(category)

  for img in os.listdir(path):
    imgpath = os.path.join(path,img)
    object_img = cv2.imread(imgpath,0)
    object_img = cv2.resize(object_img,(50,50))
    image = np.array(object_img).flatten()

    data.append([image,label])

```

```
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
import random
from sklearn.model_selection import train_test_split
```

```
random.shuffle(data)
features =[]
label1 = [] 

for feature, label in data:
  features.append(feature)
  label1.append(label)
```

```
import pandas as pd
df = pd.DataFrame(features)
df['label']=label1
df.head()

```

```
x_train,x_test,y_train,y_test = train_test_split(features,label1,test_size = 0.25)
```

```
from sklearn.svm import SVC
model = SVC(C=1.0,kernel = 'poly', gamma ='auto')
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```
y_test
```

```
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
